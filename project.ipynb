```python
import pandas as pd
```

```python
dt=pd.read_csv("rainfall.csv")
dt
```

```python
dt.isna().sum()
```

```python
dt.columns
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
dt['weather_condition']=l.fit_transform(dt['weather_condition'])
dt['weather_condition']                                        
```

```python
x=['rainfall','temperature','humidity','wind_speed','weather_condition']
for col in x:
    dt[col].fillna(dt[col].mean(),inplace=True)
dt    
```

```python
import  matplotlib.pyplot as plt
import seaborn as sns
```

```python
temp=dt[dt['tmperature']==17]
temp
```

```python
plt.plot(dt['tmperature'],dt['rainfall'],'.')
plt.title('plot of temperature and rainfall')# title of the graph
plt.xlabel('temp')# naming the x axis
plt.ylabel('rainfall')
plt.grid()
plt.show()
```

```python
plt.bar(dt['tmperature'],dt['rainfall'])
plt.title('relationship btw temp and rainfall')
plt.xlabel('temp')
plt.ylabel('rainfall')
plt.show()
```
